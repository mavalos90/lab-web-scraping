#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page

````python
# This is the url you will scrape in this exercise
url ='https://en.wikipedia.org/wiki/Python' 
````

````python
#your code
def get_links(url):
    html = requests.get(url).content
    soup = BeautifulSoup(html)
    links = str(soup.find_all('ul'))
    reference = re.findall('(?:<a href=\".)((\w+\W\w+){1,})(?:\")', ''.join(links))
    # It also captures as a second group an incomplete string
    references = [i for i,j in reference]
    return references
````

````python
links = get_links(url)
result = [i for i in links if 'Python' in i]
result
````

````python
#get a list of all the url from a random result
url = 'https://en.wikipedia.org/'+result[0]
get_links(url)
````

####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code
html = requests.get(url).content
soup = BeautifulSoup(html, 'lxml')
earthquakes = soup.find('tbody', {'id': 'tbody'}).find_all("tr")
earthquakes
````

````python
#filter text
rows = [row.text.strip().split('\n') for row in earthquakes]
rows = [list(filter(None,row)) for row in rows]
rows = [row.text.strip('\n') for row in earthquakes]
rows = rows[0:4]+rows[7:]
rows = [row.text.strip().split('\xa0') for row in earthquakes]
rows = rows[0:4]+rows[7:]
rows
````

````python
# clean each column

date = [row[0][-10:] for row in rows]
time = [row[3][:12] for row in rows]
latitude = [row[3][-5:]+row[4] for row in rows]
longitude = [row[6]+row[7] for row in rows]
region_name = [row[-1][0:-16] for row in rows]

#create the dataframe
datos = [date, time, latitude, longitude, region_name]
colnames = ['date', 'time', 'latitude', 'longitude', 'region_name']

df = pd.DataFrame(datos).T
df.columns = colnames
df
````

#### List all language names and number of related articles in the order they appear in wikipedia.org

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
#your code
html = requests.get(url).content
soup = BeautifulSoup(html, 'lxml')
extract = str(soup.find('div', {'class': 'central-featured'}))
extract
````

````python
# list of languages
languages = re.findall(r'(?:strong>)\w+', extract)
languages = [l[7:] for l in languages]
languages
````

````python
#list of articles
articles = re.findall(r'>\d+\s+\d+\s\d+', extract)
articles = [''.join(a.strip('>').split('\xa0')) for a in articles]
articles
````

````python
result = [[languages[i], articles[i]] for i in range(len(languages))]
result
````
